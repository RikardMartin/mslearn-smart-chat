```python
### Connect to Azure AI Search ###

from azure.core.credentials import AzureKeyCredential
from azure.search.documents import SearchClient

search_client = SearchClient(AI_SEARCH_ENDPOINT, AI_SEARCH_INDEX_NAME, AzureKeyCredential(AI_SEARCH_API_KEY))
search_client
```

```python
### Find and prep documents for indexing ###

folder = os.listdir(base_path)
files = [file for file in os.listdir(base_path) if '.json' in file]
files

fields = ["uid", "url", "title", "levels", "roles", "products", "subjects", "duration_in_minutes", "rating", "pupularity", "last_modified", "summary", "content", "content_vector"]

documents = []
for doc_name in files:
    new_data = {}
    with open(os.path.join(base_path, doc_name), 'r') as document:
        data = json.load(document)
        for field in data:
            if field in fields:
                new_data[field] = data[field]
            if field=="rating":
                new_data[field].pop("count")
            if field=="uid":
                new_data[field] = base64.b64encode(new_data["uid"].encode("utf-8"))

        documents.append(new_data)
        
print(len(documents), "documents processed")
```

```python
### Push documents to index ###

step = 500
for i in range(0, len(documents), step):
    chunk = documents[i:i+step]

    result = search_client.upload_documents(documents=chunk)
    print("Upload of new document succeeded: {}".format(result[0].succeeded))
```
